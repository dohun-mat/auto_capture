```python
import pyautogui
import fitz
import time
import io
import img2pdf  # 추가된 부분

time.sleep(10)

pdf_path = '~~.pdf'

start_page = 1
end_page = 200
page_transition_delay = 1

doc = fitz.open()

for page_num in range(start_page, end_page + 1):
    print(page_num)
    
    time.sleep(3) 
    pyautogui.moveTo(2404,814)
    pyautogui.click()  
#     pyautogui.press('right')
#     time.sleep(page_transition_delay)

    image = pyautogui.screenshot()
#     image.show()
    
    buffer = io.BytesIO()
    image.save(buffer, format='PNG')
    buffer.seek(0)

    pdf_bytes = img2pdf.convert(buffer.getvalue())  # 추가된 부분
    img_pdf = fitz.open("pdf", pdf_bytes)  # 수정된 부분
    rect = img_pdf[0].rect

    page = doc.new_page(width=rect.width, height=rect.height)
    page.show_pdf_page(rect, img_pdf, 0)

doc.save(pdf_path)
doc.close()
```
